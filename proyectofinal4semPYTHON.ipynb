```python
import pandas as pd
import json
import csv
import xml.etree.ElementTree as ET
import json
import pyodbc

# Configura tu cadena de conexión aquí
server = 'FABELA'
database = 'ProyectoFinal4Sem'
conn_str = f'DRIVER=ODBC Driver 17 for SQL Server;SERVER={server};DATABASE={database};Trusted_Connection=yes;'

def main():
    print("Bienvenido a la aplicación de manejo de datos con SQL Server:")

    while True:
        print("\nMenú:")
        print("1. Cargar datos desde archivo CSV")
        print("2. Cargar datos desde archivo XML")
        print("3. Cargar datos desde archivo JSON")
        print("4. Guardar datos a archivo CSV")
        print("5. Guardar datos a archivo XML")
        print("6. Guardar datos a archivo JSON")
        print("7. Mostrar datos desde SQL Server")
        print("8. Guardar datos a SQL Server")
        print("9. Salir")
        option = input("Seleccione una opción: ")

        if option == "1":
            csv_path = input("Ingrese la ruta del archivo CSV: ")
            load_csv(csv_path)
        elif option == "2":
            xml_path = input("Ingrese la ruta del archivo XML: ")
            load_xml(xml_path)
        elif option == "3":
            json_path = input("Ingrese la ruta del archivo JSON: ")
            load_json(json_path)
        elif option == "4":
            save_csv_path = input("Ingrese la ruta para guardar el archivo CSV: ")
            save_to_csv(save_csv_path)
        elif option == "5":
            save_xml_path = input("Ingrese la ruta para guardar el archivo XML: ")
            save_to_xml(save_xml_path)
        elif option == "6":
            save_json_path = input("Ingrese la ruta para guardar el archivo JSON: ")
            save_to_json(save_json_path)
        elif option == "7":
            display_data_from_sql()
        elif option == "8":
            save_data_to_sql()
        elif option == "9":
            break
        else:
            print("Opción no válida. Intente de nuevo.")

def load_csv(file_path):
    try:
        with open(file_path, newline='') as csvfile:
            reader = csv.reader(csvfile)
            headers = next(reader)
            data = [row for row in reader]
            print("Datos cargados exitosamente desde CSV.")
            return headers, data
    except Exception as ex:
        print(f"Error al cargar CSV: {str(ex)}")

def load_xml(file_path):
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()
        headers = [elem.tag for elem in root[0]]
        data = [[elem.text for elem in row] for row in root]
        print("Datos cargados exitosamente desde XML.")
        return headers, data
    except Exception as ex:
        print(f"Error al cargar XML: {str(ex)}")

def load_json(file_path):
    try:
        with open(file_path) as jsonfile:
            data = json.load(jsonfile)
            headers = list(data[0].keys())
            rows = [[str(val) for val in row.values()] for row in data]
            print("Datos cargados exitosamente desde JSON.")
            return headers, rows
    except Exception as ex:
        print(f"Error al cargar JSON: {str(ex)}")

def save_to_csv(file_path):
    headers, data = load_data_from_sql()
    try:
        with open(file_path, 'w', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(headers)
            writer.writerows(data)
        print("Datos guardados exitosamente en CSV.")
    except Exception as ex:
        print(f"Error al guardar CSV: {str(ex)}")

def save_to_xml(file_path):
    headers, data = load_data_from_sql()
    try:
        root = ET.Element("Data")
        for row in data:
            item = ET.SubElement(root, "Row")
            for i, value in enumerate(row):
                ET.SubElement(item, headers[i]).text = value
        tree = ET.ElementTree(root)
        tree.write(file_path)
        print("Datos guardados exitosamente en XML.")
    except Exception as ex:
        print(f"Error al guardar XML: {str(ex)}")

def save_to_json(file_path):
    headers, data = load_data_from_sql()
    try:
        json_data = []
        for row in data:
            json_data.append(dict(zip(headers, row)))
        with open(file_path, 'w') as jsonfile:
            json.dump(json_data, jsonfile, indent=4)
        print("Datos guardados exitosamente en JSON.")
    except Exception as ex:
        print(f"Error al guardar JSON: {str(ex)}")

def display_data_from_sql():
    try:
        headers, data = load_data_from_sql()
        print("\nDatos actuales desde SQL Server:")
        print("\t".join(headers))
        for row in data:
            print("\t".join(row))
    except Exception as ex:
        print(f"Error al mostrar datos desde SQL Server: {str(ex)}")

def load_data_from_sql():
    try:
        conn = pyodbc.connect(conn_str)
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM Employee")
        headers = [desc[0] for desc in cursor.description]
        data = cursor.fetchall()
        return headers, data
    except Exception as ex:
        print(f"Error al cargar datos desde SQL Server: {str(ex)}")

def save_data_to_sql():
    try:
        conn = pyodbc.connect(conn_str)
        cursor = conn.cursor()
        cursor.execute("DELETE FROM Employee")
        headers, data = load_data_from_sql()
        for row in data:
            cursor.execute("INSERT INTO Employee VALUES (" + ",".join(["?"] * len(headers)) + ")", row)
        conn.commit()
        print("Datos guardados exitosamente en SQL Server.")
    except Exception as ex:
        print(f"Error al guardar datos en SQL Server: {str(ex)}")

if __name__ == "__main__":
    main()
```
